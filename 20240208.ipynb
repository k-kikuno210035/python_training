```python
# https://www.python.jp/train/function/local_var.html

# function define
def calc(num_peach, num_orange):
  total = num_peach*100 + num_orange*40
  return total

# main
input_peach = input("input peach count >> ")
cnt_peach = int(input_peach)
input_orange = input("input orange count >> ")
cnt_orange = int(input_orange)

total_price = calc(cnt_peach, cnt_orange)
print("total_price = ", total_price)
```

```python
# https://www.python.jp/train/list/list_oper.html

list= [1]

# mode 0
def show_list():
  print(list)

# mode 1
def insert_value():
  list_length = len(list)
  print("input position (0～",list_length,")")
  input_position=input(" >> ")
  position = int(input_position)
  insert_value=input("insert value >>")
  list.insert(position, insert_value)
  print(list)

# mode 2
def update_value():
  list_length = len(list)
  print("input position (0～",list_length-1,")")
  update_position=input(" >> ")
  position = int(update_position)
  update_value=input("update value >>")
  list[position] = update_value
  print(list)

# mode 3
def delete_value():
  list_length = len(list)
  print("input position (0～",list_length-1,")")
  delete_position=input(" >> ")
  position = int(delete_position)
  del list[position]
  print(list)

# main
mode = 0

while mode != 99:
  input_mode=input("select mode (0:show / 1:insert / 2:update / 3:delete / other:quit) >>")
  mode = int(input_mode)

  if mode == 0:
    show_list()
  elif mode == 1:
    insert_value()
  elif mode == 2:
    update_value()
  elif mode == 3:
    delete_value()
  else:
    print("finish.")
    mode = 99
```

```python
# https://www.python.jp/train/dict/index.html
# https://www.python.jp/train/dict/dict_oper.html

dictionary = {}

# mode 0
def show_dictionary():
  print("dic >> ", dictionary)
  print("dic_num >> ", len(dictionary))

# mode 1
def insert_word():
  input_enWord = input("input English word >> ")
  input_jpWord = input("input Japanese word >> ")
  if input_enWord in dictionary:
    print(input_enWord, "is exist.")
  else:
    dictionary[input_enWord] = input_jpWord

  print("dic >> ", dictionary)
  print("dic_num >> ", len(dictionary))

# mode 2
def update_word():
  input_enWord=input("input English word >> ")
  input_jpWord=input("input Japanese word >> ")
  if input_enWord in dictionary:
    dictionary[input_enWord] = input_jpWord
  else:
    print(input_enWord, "is not exist.")

  print("dic >> ", dictionary)
  print("dic_num >> ", len(dictionary))

# mode 3
def delete_word():
  input_enWord=input("input English word >> ")
  if input_enWord in dictionary:
    del  dictionary[input_enWord]
  else:
    print(input_enWord, "is not exist.")

  print("dic >> ", dictionary)
  print("dic_num >> ", len(dictionary))


# main
mode = 0

while mode != 99:
  input_mode=input("select mode (0:show / 1:insert / 2:update / 3:delete / other:quit) >>")
  mode = int(input_mode)

  if mode == 0:
    show_dictionary()
  elif mode == 1:
    insert_word()
  elif mode == 2:
    update_word()
  elif mode == 3:
    delete_word()
  else:
    print("finish.")
    mode = 99
```
